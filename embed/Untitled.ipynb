```python
import numpy as np
import matplotlib.pyplot as plt

M = np.linspace(1, 7, 7)

a = 12 * (M - 1)
a2 = 12 * M - 10

plt.figure()
plt.plot(M, a)
plt.plot(M, a2)
plt.show()
```

```python


import networkx as nx
import dwave_networkx as dnx
import minorminer as mm

C = nx.complete_graph(12)

pegasus = dnx.pegasus_graph(3)

embedding = mm.find_embedding(C, pegasus)

colors = {}
for x in C.nodes:
    colors[x] = (0, 0, 1, 1)

#plt.figure(figsize=(13, 5))
plt.figure(figsize=(50, 20))
dnx.draw_pegasus_embedding(pegasus, embedding, node_size=20, chain_color=colors)
plt.show()
#plt.savefig('pegasus.pdf')
```

```python


import networkx as nx
import dwave_networkx as dnx
import minorminer as mm

C = nx.Graph()
edges = []
for i in range(1, 5):
    for j in range(5, 9):
        edges.append((i, j))
print(edges)
C.add_edges_from(edges)

pegasus = dnx.pegasus_graph(3)

embedding = mm.find_embedding(C, pegasus)

colors = {}
for x in C.nodes:
    colors[x] = (0, 0, 1, 1)

#plt.figure(figsize=(13, 5))
plt.figure(figsize=(50, 20))
dnx.draw_pegasus_embedding(pegasus, embedding, node_size=20, chain_color=colors)
plt.show()
#plt.savefig('pegasus.pdf')
```
